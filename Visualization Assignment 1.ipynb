```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("customer_churn (1).csv")
```

```python
data.head()
```

```python
data.Contract.value_counts()
```

```python
x= data['Contract'].value_counts().keys() #to get keys
```

```python
x
```

```python
y =  data['Contract'].value_counts().tolist() #to take only num of keys
```

```python
y
```

```python
plt.bar(x,y,color='orange')
plt.xlabel("Contract type of customer")
plt.ylabel("Count")
plt.title("Distribution of Contract")
plt.show()
```

```python
plt.hist(data['MonthlyCharges'],color='forestgreen')
plt.xlabel("MonthlyCharges")
plt.ylabel("Count")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
plt.hist(data['MonthlyCharges'],color='forestgreen',bins=30)
plt.xlabel("MonthlyCharges")
plt.ylabel("Count")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
plt.hist(data['MonthlyCharges'],color='forestgreen',bins=10)
plt.xlabel("MonthlyCharges")
plt.ylabel("Count")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
plt.scatter(x=data['tenure'].head(20),y=data['TotalCharges'].head(20),color='indigo')
plt.xlabel("Tenure of the customer")
plt.ylabel("Total charges in cut")
plt.title("Total charges vs Tenure")
plt.show()
```

```python
plt.scatter(x=data['tenure'],y=data['TotalCharges'],color='indigo')
plt.xlabel("Tenure of the customer")
plt.ylabel("Total charges in cut")
plt.title("Total charges vs Tenure")
plt.show()
```

```python
data.boxplot(column="MonthlyCharges",by=["PaymentMethod"],figsize=(10,10),color="olive")
plt.xlabel("Payment Method of Customer")
plt.ylabel("Monthly Charges Incurred")
plt.title("Monthly Charges vs Payment Method")
```

```python
import seaborn as sns
sns.boxplot(x=data['PaymentMethod'],y=data['MonthlyCharges'])
```
